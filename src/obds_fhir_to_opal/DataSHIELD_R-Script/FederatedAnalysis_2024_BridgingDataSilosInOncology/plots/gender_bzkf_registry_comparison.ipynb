```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_bar_chart():
    categories = [
        "Breast (C50, D05)",
        "Thyroid (C73)",
        "Hodgkin Lymphoma (C81)",
        "Pancreas (C25)",
        "Gallbladder and Biliary Tract (C23-C24)",
        "Larynx (C32)",
        "Esophagus (C15)",
        "Bladder (C67, D09.0, D41.1)",
        "Liver (C22)",
        "Kidney (C64)",
    ]
    female_bzkf = [99, 67, 49, 48, 46, 15, 21, 25, 26, 29]
    female_registry = [99, 69, 45, 48, 44, 16, 19, 23, 29, 30]
    male_bzkf = [1, 33, 51, 52, 54, 85, 79, 75, 74, 71]
    male_registry = [1, 31, 55, 52, 56, 84, 81, 77, 71, 70]

    x = np.arange(len(categories))

    width = 0.2

    fig, ax = plt.subplots(figsize=(10, 6))

    female_bzkf_bars = ax.bar(
        x - 1.5 * width, female_bzkf, width, label="Female BZKF (%)", color="#66c159"
    )
    female_registry_bars = ax.bar(
        x - 0.5 * width,
        female_registry,
        width,
        label="Female Registry (%)",
        color="lightgreen",
    )
    male_bzkf_bars = ax.bar(
        x + 0.5 * width, male_bzkf, width, label="Male BZKF (%)", color="#7187da"
    )
    male_registry_bars = ax.bar(
        x + 1.5 * width,
        male_registry,
        width,
        label="Male Registry (%)",
        color="lightblue",
    )

    for bars, data in zip(
        [female_bzkf_bars, female_registry_bars, male_bzkf_bars, male_registry_bars],
        [female_bzkf, female_registry, male_bzkf, male_registry],
    ):
        for bar, value in zip(bars, data):
            height = bar.get_height()
            ax.text(
                bar.get_x() + bar.get_width() / 2,
                height + 1,
                value,
                ha="center",
                va="bottom",
                fontsize=10,
            )

    ax.set_ylabel("Percentage (%)")
    ax.set_xticks(x)
    ax.set_xticklabels(categories, rotation=45, ha="right")

    ax.legend()

    plt.tight_layout()
    plt.show()


plot_bar_chart()
```
